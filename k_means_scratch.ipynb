```python
file_path = r"C:\Users\heumo\Machine learning\student_clustering (1).csv"
#df = np.genfromtxt(file_path, delimiter=',')
df = pd.read_csv(file_path)
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(df)
    wcss.append(km.inertia_)
wcss
```

```python
import matplotlib.pyplot as plt

# Votre code pour calculer wcss

plt.plot(range(1, 11), wcss, marker='o')  # Tracé de la courbe
plt.title('Elbow Method')  # Titre du graphe
plt.xlabel('Number of clusters')  # Légende de l'axe des x
plt.ylabel('WCSS')  # Légende de l'axe des y
plt.show()  # Affichage du graphe
```

```python
X = df.iloc[:,:].values
km = KMeans(n_clusters=4)
y_means = km.fit_predict(X)
```

## 3d Data

```python
# Définir les centres comme un tableau numpy
centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
# Définir les déviations standard pour chaque blob
cluster_std = [1,1,1,1]

# Générer un ensemble de données avec 200 échantillons, des centres spécifiés, des déviations standard et 3 caractéristiques
X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
X
```

```python
# Créer une figure et un axe 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracer les points en 3D
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', marker='o', edgecolors='black')

# Définir les labels des axes
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

# Afficher le graphique
plt.title('Scatter Plot 3D')
plt.show()
```
